```python
import pandas as pd
import matplotlib.pyplot as plt
from  math import *
import numpy as np
import random
```

```python
#Following function will create rarefaction curve given input of list of individuals and number of individuals taken in single event
#
#For the following 'rarefaction' function 'individuals' is a list of individuals, samples is number of individuals taken
#in one event
#
def rarefaction(individuals,samples):
    n = len(individuals)/samples
    split = np.array_split(individuals,n) #This will split randomised list in sublists in which, a sublist will act as one event of random sampling
    
    l2 = []
    unique_species = []
    for i in split: #This will count the no. of unique species found in each list and append it in unique_species
        count1=0
        for j in i:
            if j not in l2:
                count1+=1
                l2.append(j)
        unique_species.append(count1)

    cumsum = []  #cumulitive sum for new species found in each draw.
    x=0
    for i in unique_species:
        #x=0
        x= x+i
        cumsum.append(x)
        
    total_ind = [i*samples for i in range(1,len(unique_species)+1)]
    plt.plot(total_ind, cumsum)
    plt.xlabel("Total Individuals")
    plt.ylabel("Species")
    plt.title("Rarefaction Curve")
    
###########################################################################################################################
############################################# END OF THE FUNCTION #########################################################
###########################################################################################################################

data = pd.read_excel("Problem set-1.xls")#data is read here and stored in 'data'
ind = data["Individuals"] #individuals are stored in 'ind'
ad = np.array(ind) #list of individuals is created here
ad1=ad[:-2]


individuals=[]
y=1
for i in ad1:  #This will create list of individual, each individual is labelled by it's species number.
    x=1
    while x!=i+1:
        individuals.append(y)
        x=x+1
    y=y+1


#random.shuffle(individuals)  #This function will randomly shuffle list of individuals.

#rarefaction(individuals,50)
for i in range(1000):
    random.shuffle(individuals)
    rarefaction(individuals,50)
```
